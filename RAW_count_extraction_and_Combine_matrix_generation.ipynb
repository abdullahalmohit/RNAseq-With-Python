Extracting Raw Counts from the featureCounts Output

```python
import pandas as pd

# Load the featureCounts output
fc_df = pd.read_csv("/content/SRR29685140_counts.txt", sep='\t', comment='#')

# Extract raw counts: gene ID + count columns (assume first 6 are annotations)
raw_counts = fc_df.iloc[:, [0] + list(range(6, fc_df.shape[1]))]

# Set gene IDs as index
raw_counts.set_index('Geneid', inplace=True)

# Save to CSV
raw_counts.to_csv("SRR29685140_raw_counts_matrix.csv")

print("Raw count matrix saved as 'SRR29685140_raw_counts_matrix.csv'")
```

RAW count matrices to Combined RAW count matrix

```python
import pandas as pd
import glob

# Find all raw count files in the current folder
files = glob.glob("*_raw_counts_matrix.csv") #Finds all filenames that match the pattern "*_raw_counts_matrix.csv"

# Read each file into a dataframe
dfs = [pd.read_csv(f, index_col=0) for f in files]

# Merge by gene ID (row names)
combined = pd.concat(dfs, axis=1)

# Save the combined matrix
combined.to_csv("Combined_RAW_Counts.csv")

print("Combined raw count matrix saved as 'Combined_RAW_Counts.csv'")
```
